Descarrega el data set Airlines Delay: Airline on-time statistics and delay causes i carrega’l a un Pandas Dataframe. Explora les dades que conté, explica breument quines variables hi ha i queda’t únicament amb les columnes que consideris rellevants. Justifica la teva elecció.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
airlines_delay = pd.read_csv('DelayedFlights.csv')
```

```python
df = airlines_delay.sample(n=20000)
```

Aquest Data Set analitza 1.936.758 vols interns diferents als EUA durant l'any 2008 i les seves causes de retard, desviament i cancel·lació, si n'hi ha. En el nostre anàlisi agafarem una mostra aleatòria de 200.000 registres. Hi ha 30 columnes i ens en quedem amb 21. S'ha eliminat l'index autogenerat, totes les columnes referides als taxis, el codi de cancelació dels vols i els motius del retard, ja que és informació específica sense la qual es pot fer l'anàlisi sol·licitat. 

```python
df.info()
```

- Unnamed: index autogenerat
- Year: 2008
- Month: 1-12
- DayofMonth: 1-31
- DayOfWeek: 1 (dilluns) - 7 (diumenge)
- DepTime: hora de sortida real(local, hhmm)
- CRSDepTime: Hora de sortida programada(local, hhmm)
- ArrTime: hora d'arribada real(local, hhmm)
- CRSArrTime: hora d'arribada programada(local, hhmm)
- UniqueCarrier: Codi únic de transportista
- FlightNum: número de vol
- TailNum: número de cua de l'avió, matrícula de l'aeronau, identificador únic de l'avió
- ActualElapsedTime: en minuts
- CRSElapsedTime: en minuts
- AirTime: en minuts
- ArrDelay: en minuts. Un vol es compta com a "a temps" si el retard és de menys de 15 minuts. 
- DepDelay: retard de sortida, en minuts
- Origin: codi d'aeroport IATA
- Dest: Codi de l'aeroport de destinació IATA
- Distance: Distància en milles
- TaxiIn: taxi que va puntual, en minuts
- TaxiOut: taxi que va amb retard, en minuts
- Cancelled: el vol s'ha cancel·lat
- CancellationCode: motiu de la cancel·lació (A = operador, B = temps, C = NAS, D = seguretat)
- Diverted: 1 = sí, 0 = no
- CarrierDelay: en minuts, el retard té a veure amb el transport aeri i les diferents casuistiques
- WeatherDelay: en minuts, el retard meteorològic és a causa de previsió o certesa de condicions metereologiques extremes. 
- NASDelay: en minuts, el retard que està sota el control del Sistema Nacional d'Espai Aeri (NAS) pot incloure: condicions meteorològiques no extremes, operacions aeroportuàries, volum de trànsit intens, control de trànsit aeri, etc.
- SecurityDelay: retard en minuts, pot ser causat per qualsevol problema de seguretat detectat 
- LateAircraftDelay en minuts: retard a l'arribada a un aeroport a causa de l'arribada tardana de la mateixa aeronau a un aeroport anterior. L'efecte dominó d'un retard anterior als aeroports aigües avall es coneix com a propagació del retard.

```python
DelFlights = df.drop(['Unnamed: 0','TaxiIn', 'TaxiOut','CancellationCode','CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay'], axis=1)
```

```python
# Ranking dels dies de la setmana en relació als retards, primer busquem els dies que realment hi hagi retard(= o + de 15 minuts) i fem una nova columna. 

DelFlights['Delayed'] = df['ArrDelay'].where(DelFlights['ArrDelay'] >= 15)

DelFlights.groupby(['DayOfWeek'])['Delayed'].count().sort_values(ascending=False)
```

```python
# Ranking dels mesos de l'any en relació als retards

DelFlights.groupby(['Month'])['Delayed'].count().sort_values(ascending=False)
```

```python
# Ranking dels dies dels mesos de l'any en relació als retards

DelFlights.groupby(['DayofMonth'])['Delayed'].count().sort_values(ascending=False)
```

```python
# Mostrem els vols cancel·lats

CancelFlights = DelFlights.loc[DelFlights['Cancelled'] > 0]
CancelFlights
```

```python
# Mostrem els vols desviats 

DivertedFlights = DelFlights.loc[DelFlights['Diverted'] > 0]
DivertedFlights
```

#### Troba quantes dades faltants hi ha per columna.

```python
missing_values_count = DelFlights.isna().sum()

print(missing_values_count)
```

```python
DelFlights = DelFlights.fillna(0)
```

#### Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

```python
# Per la velocitat mitjana de vol convertim les Milles en Km i els minuts en hores, ho dividim i treiem la mitjana.

DelFlights['DistanceKm'] = DelFlights['Distance']* 1.60934
DelFlights['TimeHour'] = DelFlights['AirTime']/60
DelFlights['MeanFlightSpeed'] = DelFlights['DistanceKm']/DelFlights['TimeHour'].mean()

DelFlights.head()
```

```python
# Només es compten com a retard els vols que arriben més de 15 minuts tard

DelFlights['DelayedOrNot'] = np.where(DelFlights['ArrDelay'] >= 15, 'Delayed', 'On time')
```

```python
DelFlights[['ArrDelay','DelayedOrNot']]
```

#### Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards. Aporta allò que consideris rellevant.

```python
# Vols més llargs
DelFlights.sort_values(by = 'Distance', ascending = False)
```

```python
# Vols més endarrerits

DelFlights.sort_values(by = 'Delayed', ascending = False)
```

```python
# Rutes més llargues 

DelFlights_grouped = DelFlights.groupby(['Origin','Dest']).agg({'Distance':'max'})

DelFlights_grouped = DelFlights_grouped.sort_values(by='Distance', ascending=False)

DelFlights_grouped.head(20)
```

```python
# Rutes més curtes

DelFlights_grouped = DelFlights.groupby(['Origin','Dest']).agg({'Distance':'max'})

DelFlights_grouped = DelFlights_grouped.sort_values(by='Distance', ascending=True)

DelFlights_grouped.head(20)
```

```python
# Avions amb origens i destinacions amb menys distància i major retard

origin_dest_distance_delay = DelFlights[['TailNum','Origin','Dest','Distance', 'ArrDelay']]
```

```python
origin_dest_distance_delay.sort_values(by = ['Distance', 'ArrDelay'], ascending = [True, False]) 
```
